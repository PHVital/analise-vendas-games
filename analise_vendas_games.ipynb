## 2. Limpeza e Pré-processamento de Dados
Nesta seção, realizaremos as etapas necessárias para limpar e preparar o dataset para análise, incluindo a renomeação de colunas, tratamento de tipos de dados e a gestão de valores ausentes.

### 2.1 Renomeação de Colunas
Para facilitar a manipulação dos dados no Python e seguir as convenções de nomenclatura (snake_case), as colunas do DataFrame serão renomeadas.

```python
print("Nomes das colunas antes da renomeação:")
print(df_games.columns)
df_games.columns = df_games.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
print("\nNomes das colunas após a renomeação (snake_case):")
print(df_games.columns)
```

### 2.2 Verificação e Ajuste de Tipos de Dados
É importante garantir que cada coluna esteja com o tipo de dado apropriado para as análises. A coluna 'year', por exemplo, foi carregada inicialmente como float e precisa ser convertida para um tipo inteiro que suporte valores ausentes (Int64).

```python
print(f"Tipo da coluna 'Year': {df_games['year'].dtype}")
```

```python
try:
    df_games['Year'] = df_games['Year'].astype('Int64')
    print("Coluna 'Year' convertida para Int64.")
except Exception as e:
    print(f'Não foi possível converter "Year" para Int64: {e}')
```

### 2.3 Tratamento de Valores Ausentes
Após a inspeção inicial, identificamos valores ausentes nas colunas 'year' e 'publisher'. A seguir, detalhamos a contagem e a estratégia adotada para cada uma.

```python
print("\nContagem total de valores ausentes por coluna (APÓS tratamentos iniciais):")
print(df_games.isnull().sum())
```

Para a coluna 'year', que apresentava 271 valores ausentes (aproximadamente 1.6% do dataset), e sendo o ano de lançamento uma informação crucial para análises de tendência, optou-se por remover as linhas onde esta informação estava faltando.

Para a coluna 'publisher', com 58 valores ausentes (aproximadamente 0.35% do dataset), também optamos por remover essas linhas para manter a consistência dos dados para análise de publicadoras. Uma alternativa seria preenchê-los com "Unknown", mas dado o baixo volume, a remoção foi considerada aceitável para este estudo.

```python
df_games.dropna(subset=['year'], inplace=True)
df_games.dropna(subset=['publisher'], inplace=True)

print("\nContagem total de valores ausentes por coluna (APÓS tratamentos de valores ausentes):")
print(df_games.isnull().sum())
```

```python
print(f"\nNúmero de linhas duplicadas: {df_games.duplicated().sum()}")
```

## 3. Análise Exploratória de Dados (EDA)

Com os dados limpos e preparados, podemos iniciar a análise exploratória para entender melhor as características e tendências do mercado de video games.

### 3.1 Análise Univariada

Começaremos analisando as variáveis individualmente para entender suas distribuições e principais características.

#### 3.1.1 Distribuição das Vendas Globais por Jogo

Para entender o desempenho comercial dos jogos, vamos analisar a distribuição da variável `global_sales`. Um histograma nos mostrará a frequência de diferentes faixas de vendas, enquanto um boxplot ajudará a identificar a mediana, os quartis e potenciais outliers.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_games[df_games['global_sales'] < 10]['global_sales'], bins=50, kde=True)
plt.title('Distribuição das Vendas Globais dos Jogos')
plt.xlabel('Vendas Globais (em milhões)')
plt.ylabel('Frequência (Número de Jogos)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_games['global_sales'])
plt.title('Boxplot das Vendas Globais')
plt.xlabel('Vendas Globais (em milhões)')
plt.show()
```

**Observações sobre as Vendas Globais:**

* **Concentração em Vendas Baixas:** O histograma (especialmente se não filtrado) demonstra claramente que a grande maioria dos jogos no dataset possui um volume de vendas globais relativamente baixo (provavelmente abaixo de 1 ou 2 milhões de unidades). A distribuição é fortemente assimétrica à direita.
* **Outliers de Sucesso:** O boxplot evidencia a presença de muitos outliers, que representam os jogos de grande sucesso comercial (blockbusters) com vendas significativamente mais altas que a média. Esses poucos títulos de alto desempenho puxam a média de vendas para cima, enquanto a mediana (representada pela linha central no boxplot) provavelmente está em um valor mais baixo, refletindo melhor o desempenho do "jogo típico".
* **Cauda Longa:** A "cauda longa" da distribuição indica que, embora raros, existem jogos que alcançam volumes de vendas extraordinariamente altos, como "Wii Sports" que lidera o ranking.

Essa distribuição é comum em mercados de entretenimento, onde poucos produtos se tornam grandes sucessos e muitos outros têm um desempenho mais modesto.

#### 3.1.2 Evolução do Número de Lançamentos de Jogos por Ano

Analisar o número de jogos lançados a cada ano pode nos dar insights sobre o crescimento da indústria, o ciclo de vida de plataformas e possíveis impactos de eventos externos.

```python
plt.figure(figsize=(15,7))
games_per_year = df_games.groupby('year')['name'].count()
sns.lineplot(data=games_per_year)
plt.title('Número de Jogos Lançados por Ano')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Número de Jogos Lançados')
plt.xticks(rotation=45)
plt.show()
```

**Tendências nos Lançamentos Anuais:**

* **Crescimento da Indústria:** O gráfico de linha ilustra um crescimento expressivo no número de jogos lançados a partir do final da década de 1990, atingindo um pico entre os anos de 2008 e 2009, com mais de 1400 títulos lançados anualmente. Esse período coincide com o auge de consoles como o Nintendo Wii, PlayStation 3 e Xbox 360.
* **Queda nos Anos Seguintes:** Após o pico, observa-se uma tendência de queda no número de lançamentos, que se acentua após 2011.
* **Importante Limitação do Dataset:** É crucial notar que a queda drástica no número de títulos registrados no dataset após 2016 (com apenas 3 jogos em 2017 e 1 em 2020, conforme verificado na etapa de limpeza) é, muito provavelmente, um artefato da coleta de dados e da data em que o dataset foi compilado, e **não reflete uma interrupção real tão severa na indústria de games nesses anos.** Portanto, análises de tendência que incluem o período pós-2016 devem ser interpretadas com extrema cautela. O dataset parece ser mais completo e representativo para o período até aproximadamente 2015/2016.

#### 3.1.3 Distribuição de Jogos por Gênero

Vamos identificar quais gêneros de jogos são os mais prolíficos em termos de número de títulos lançados.

```python
plt.figure(figsize=(12, 8))
sns.countplot(y=df_games['genre'], order = df_games['genre'].value_counts().index)
plt.title('Número de Jogos por Gênero')
plt.xlabel('Número de Jogos')
plt.ylabel('Gênero')
plt.show()
```

* Gêneros como "Puzzle" e "Strategy", embora clássicos, aparecem com um volume menor de títulos, o que pode indicar nichos de mercado específicos ou uma menor frequência de grandes lançamentos em comparação com os gêneros de ação e esportes.

#### 3.1.4 Análise das Principais Publicadoras (Publishers)

Investigaremos quais empresas (publishers) se destacam no mercado, tanto pelo volume de jogos lançados quanto pelo total de vendas globais geradas.

```python
top_publishers_count = df_games['publisher'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
df_top_publishers_count = top_publishers_count.reset_index()
sns.barplot(
    data=df_top_publishers_count,
    x=top_publishers_count.values, 
    y=top_publishers_count.index,
    hue='publisher',
    dodge=False,
    palette='mako',
    legend=False
)
plt.title('Top 10 Publishers por Número de Jogos Lançados')
plt.xlabel('Número de Jogos')
plt.ylabel('Publisher')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_publishers_sales = df_games.groupby('publisher')['global_sales'].sum().nlargest(10).reset_index()
top_publishers_sales.columns = ['publisher', 'global_sales']
sns.barplot(
    data=top_publishers_sales,
    x='global_sales',
    y='publisher',
    hue='publisher',
    dodge=False,
    palette="rocket",
    legend=False
)
plt.title('Top 10 Publishers por Vendas Globais Totais')
plt.xlabel('Vendas Globais Totais (em milhões)')
plt.ylabel('Publisher')
plt.show()
```

**Destaques entre as Publicadoras:**

* **Volume de Lançamentos:** Empresas como "Electronic Arts", "Activision", e "Namco Bandai Games" lideram em termos de quantidade de jogos publicados, indicando uma ampla presença no mercado.
* **Liderança em Vendas:** Quando analisamos o total de vendas globais, a "Nintendo" se destaca significativamente, seguida por "Electronic Arts" e "Activision". Isso sugere que, embora a Nintendo possa não ter o maior volume de títulos de terceiros em todas as plataformas, seus próprios títulos e franquias têm um impacto comercial imenso.
* **Eficiência vs. Volume:** Comparar os dois rankings pode dar insights sobre a "eficiência" de vendas por jogo de cada publisher. Uma publisher pode ter muitos jogos (alto no primeiro ranking) mas não necessariamente estar no topo de vendas totais se muitos desses jogos forem de nicho ou tiverem vendas modestas.

#### 3.1.5 Análise das Principais Plataformas

Assim como com as publicadoras, vamos identificar as plataformas (consoles e PC) que tiveram mais jogos lançados e que geraram maior volume de vendas.

```python
top_platform_count = df_games['platform'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
df_top_platform_count = top_platform_count.reset_index()
sns.barplot(
    data=df_top_platform_count,
    x=top_platform_count.values, 
    y=top_platform_count.index,
    hue='platform',
    dodge=False,
    palette='mako',
    legend=False
)
plt.title('Top 10 Platforms por Número de Jogos Lançados')
plt.xlabel('Número de Jogos')
plt.ylabel('Platforms')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_platform_sales = df_games.groupby('platform')['global_sales'].sum().nlargest(10).reset_index()
top_platform_sales.columns = ['platform', 'global_sales']
sns.barplot(
    data=top_platform_sales,
    x='global_sales',
    y='platform',
    hue='platform',
    dodge=False,
    palette="rocket",
    legend=False
)
plt.title('Top 10 Platforms por Vendas Globais Totais')
plt.xlabel('Vendas Globais Totais (em milhões)')
plt.ylabel('platform')
plt.show()
```

**Plataformas de Destaque:**

* **Sucesso Comercial:** Em termos de vendas globais totais, o "PS2" lidera, seguido pelo "Xbox 360", "PS3" e "Wii". Isso mostra o impacto comercial dessas gerações de consoles.
* **Ciclos de Vida:** A análise por plataforma pode, indiretamente, refletir os ciclos de vida dos consoles, com plataformas mais antigas tendo acumulado mais títulos e vendas ao longo do tempo (dentro do período coberto pelo dataset).

### 3.2 Análise Bivariada e Multivariada

Nesta seção, exploraremos as relações entre duas ou mais variáveis para descobrir padrões mais complexos e insights sobre o mercado de video games.

#### 3.2.1 Desempenho Comercial Médio dos Gêneros

Após analisarmos a popularidade dos gêneros pelo número de títulos lançados, é interessante investigar o desempenho comercial médio de cada gênero. Isso pode revelar se gêneros com menos títulos conseguem, em média, um sucesso de vendas maior.

```python
plt.figure(figsize=(12, 8))
avg_sales_by_genre = df_games.groupby('genre')['global_sales'].mean().sort_values(ascending=False)
df_avg_sales = avg_sales_by_genre.reset_index()
sns.barplot(
    data=df_avg_sales,
    x=avg_sales_by_genre.values, 
    y=avg_sales_by_genre.index, 
    hue='genre',
    dodge=False,
    palette="viridis",
    legend=False
)
plt.title('Vendas Médias Globais por Gênero')
plt.xlabel('Vendas Médias Globais (em milhões)')
plt.ylabel('Gênero')
plt.show()
```

**Análise das Vendas Médias por Gênero:**

* O gráfico de barras indica que o gênero "Platform" (Plataforma) possui a maior média de vendas globais por título, seguido por "Shooter" (Tiro) e "Role-Playing" (RPG).
* Isso é particularmente interessante quando comparado à contagem de jogos por gênero. Enquanto "Action" e "Sports" lideram em volume de lançamentos, sua média de vendas por jogo é menor que a dos top 3 gêneros por média de vendas.
* Isso pode sugerir que, embora menos numerosos, os jogos de Plataforma, Tiro e RPG que são lançados tendem a ter um impacto comercial individual maior, ou que esses gêneros contêm alguns "blockbusters" que elevam significativamente a média. Gêneros como "Adventure", por outro lado, apesar de terem um número razoável de títulos, apresentam a menor média de vendas globais.

#### 3.2.2 Desempenho Comercial Médio das Plataformas e das Publishers

Também seria interessante analisar o desempenho médio das Plataformas e das Publishers, para podermos cruzar os dados.

```python
media_plataforma = df_games.groupby('platform')['global_sales'].mean().sort_values(ascending=False)

media_publisher = df_games.groupby('publisher')['global_sales'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=media_plataforma.values[:10], 
    y=media_plataforma.index[:10], 
    hue=media_plataforma.values[:10],
    palette='viridis',
    legend=False
)
plt.title('Top 10 Plataformas por Média de Vendas por Jogo')
plt.xlabel('Média de Vendas Globais (em milhões)')
plt.ylabel('Plataforma')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(
    x=media_publisher.values, 
    y=media_publisher.index, 
    hue=media_publisher.values,
    palette='plasma',
    legend=False
)
plt.title('Top 10 Publishers por Média de Vendas por Jogo')
plt.xlabel('Média de Vendas Globais (em milhões)')
plt.ylabel('Publisher')
plt.show()
```

#### 3.2.4 Correlação entre Vendas nas Diferentes Regiões

Para entender como as vendas em uma região se relacionam com as vendas em outras e com o total global, calcularemos e visualizaremos a matriz de correlação. Isso nos ajudará a identificar se o sucesso em um mercado tende a se refletir em outros.

```python
sales_columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'global_sales']
correlation_matrix = df_games[sales_columns].corr()
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor da Correlação entre Vendas Regionais e Globais')
plt.show()
```

**Insights da Matriz de Correlação de Vendas:**

* **Forte Correlação com Vendas Globais:** Observa-se uma correlação positiva muito forte (próxima de 1) entre as vendas em cada uma das principais regiões (NA, EU, JP) e as vendas globais (`global_sales`). Isso é esperado, já que as vendas globais são, em grande parte, a soma das vendas regionais. A América do Norte (`na_sales`) apresenta a correlação mais alta com as vendas globais, seguida pela Europa (`eu_sales`).
* **Correlação entre NA e EU:** Há também uma correlação positiva considerável entre as vendas na América do Norte e na Europa (ex: ~0.77), sugerindo que jogos que vendem bem em uma dessas regiões tendem a vender bem na outra, indicando preferências de mercado possivelmente similares para muitos títulos.
* **Menor Correlação com o Japão:** As vendas no Japão (`jp_sales`) mostram uma correlação mais baixa com as vendas na América do Norte e Europa, o que pode indicar um mercado com preferências mais distintas e menos influenciado diretamente pelo desempenho em mercados ocidentais, e vice-versa.
* **Outras Regiões:** A coluna `other_sales` também exibe correlações positivas com as outras regiões e com as vendas globais, embora geralmente com magnitudes um pouco menores que NA e EU.

Essas correlações são úteis para entender a interdependência dos mercados e podem informar estratégias de lançamento e marketing global.

#### 3.2.3 Preferências Regionais de Gêneros de Jogos

Vamos agora detalhar as vendas totais de cada gênero nas principais regiões de mercado (América do Norte, Europa, Japão e Outras Regiões) para identificar possíveis preferências regionais.

```python
genre_sales_region = df_games.groupby('genre')[[
    'na_sales',
    'eu_sales',
    'jp_sales',
    'other_sales'
]].sum()
genre_sales_region.sort_values(by='na_sales', ascending=False).plot(kind='bar', figsize=(15, 8), stacked=True)
plt.title('Vendas Totais por Gênero em Diferentes Regiões')
plt.xlabel('Genêro')
plt.ylabel('Vendas Totais (em milhões)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Região')
plt.tight_layout()
plt.show()
```

**Análise Regional de Vendas por Gênero:**

O gráfico de barras empilhadas (ou agrupadas) acima ilustra a contribuição de cada região principal para o total de vendas de cada gênero de jogo.

**Observações Detalhadas:**
* **Domínio da América do Norte (NA):** Em gêneros de grande volume como "Action", "Sports", e especialmente "Shooter", a América do Norte consistentemente representa a maior fatia das vendas. Isso pode ser atribuído ao tamanho do mercado e à popularidade histórica desses gêneros na região.
* **Preferências Marcantes no Japão (JP):** O mercado japonês demonstra uma forte e distinta preferência por jogos do gênero "Role-Playing" (RPG), onde sua contribuição para as vendas globais do gênero é significativamente maior do que em outros. Em contraste, gêneros como "Shooter" têm uma penetração de mercado notavelmente menor no Japão.
* **Europa (EU) como Mercado Chave:** A Europa frequentemente se posiciona como o segundo maior mercado para muitos gêneros, seguindo padrões de preferência relativamente similares aos da América do Norte, embora com volumes totais geralmente menores.

#### 3.2.4 Ciclos de Vida e Dominância de Plataformas ao Longo do Tempo

Analisar a evolução anual das vendas globais para plataformas selecionadas nos permite visualizar seus ciclos de vida, períodos de dominância e o impacto da chegada de novas gerações de consoles.

```python
platforms_to_analyze = ['PS2', 'X360', 'DS', 'PC', 'PS4']
platform_sales_year = (
    df_games[df_games['platform'].isin(platforms_to_analyze)]
        .groupby(['year', 'platform'])['global_sales']
        .sum()
        .reset_index()
)

plt.figure(figsize=(15, 8))
sns.lineplot(
    data=platform_sales_year,
    x='year',
    y='global_sales',
    hue='platform',
    marker='o'
)
plt.title('Evolução Anual das Vendas Globais por Plataforma Selecionada')
plt.xlabel('Ano')
plt.ylabel('Vendas Globais (em milhões)')
plt.legend(title='Plataforma', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Ciclos de Vida das Plataformas:**

O gráfico de linha mostra a trajetória de vendas anuais para algumas das plataformas mais significativas do dataset. Podemos observar claramente:
* **Ascensão e Queda:** Cada plataforma exibe um ciclo de vida típico, com um período de crescimento nas vendas após o lançamento, um pico de popularidade e, subsequentemente, um declínio à medida que novas gerações de consoles são introduzidas e ganham tração no mercado.
* **Dominância de Gerações:** Por exemplo, o "PS2" (PlayStation 2) mostra um longo período de vendas robustas, atingindo seu pico em 2004. Plataformas como "Wii" e "Xbox 360" tiveram seus próprios períodos de grande sucesso, muitas vezes sobrepondo-se parcialmente, indicando a competição entre gerações.
* **Impacto de Novas Gerações:** A queda nas vendas de uma plataforma geralmente coincide com o lançamento e a ascensão da sua sucessora ou de consoles concorrentes de nova geração. Por exemplo, a entrada do "PS3" e "Xbox 360" impactou as vendas do "PS2".
* **Limitação para Anos Recentes:** Conforme discutido anteriormente, a aparente queda abrupta ou ausência de dados para plataformas mais recentes (como o "PS4" nos últimos anos do gráfico) deve-se à incompletude do dataset para o período pós-2016. A análise é mais robusta para os ciclos de vida completados dentro do período bem coberto pelo dataset.

## 5. Análise do Market Share Regional ao Longo dos Anos
Vamos investigar como a participação de mercado de cada região principal (América do Norte, Europa, Japão e Outras Regiões) nas vendas globais totais de video games evoluiu ano a ano. Isso pode nos mostrar mudanças nas dinâmicas de mercado e na importância relativa de cada região.

```python
sales_by_year_region = df_games.groupby('year')[
    ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
].sum()

sales_by_year_region['total_annual_sales'] = sales_by_year_region.sum(axis=1)

market_share_by_year = sales_by_year_region[
    ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
].apply(
    lambda x: (x / sales_by_year_region['total_annual_sales']) * 100, axis=0
)

plt.figure(figsize=(17,9))
market_share_by_year.plot(
    kind='area', 
    stacked=True, 
    figsize=(17, 9), 
    colormap='viridis'
)

plt.title('Participação de Mercado (%) das Vendas Globais por Região ao Longo dos Anos', fontsize=16)
plt.ylabel('Participação de Mercado (%)', fontsize=14)
plt.xlabel('Ano de Lançamento', fontsize=14)
plt.legend(title='Região', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print('\nParticipação de Mercado (%) nos últimos anos:')
print(market_share_by_year.tail(10))
```

### Dinâmica do Market Share Regional

O gráfico de área empilhada acima demonstra a evolução da participação percentual de cada grande mercado (América do Norte, Europa, Japão e Outras Regiões) no total de vendas globais de video games, ano a ano.

**Principais Observações:**
* **Domínio Histórico da América do Norte (NA):** É evidente que a América do Norte tem sido, historicamente, o maior mercado, respondendo pela maior fatia das vendas globais durante muitos anos, especialmente durante os picos de popularidade de certas gerações de consoles.
* **Crescimento e Estabilidade da Europa (EU):** O mercado europeu também se mostra consistentemente forte, muitas vezes sendo o segundo maior contribuidor para as vendas globais, com uma participação que parece ter se consolidado ao longo do tempo.
* **Importância do Japão (JP) e Suas Particularidades:** O mercado japonês, embora com uma participação geral menor que NA e EU, exibe picos de importância, especialmente em certos períodos que podem coincidir com o sucesso de consoles ou gêneros de fabricação local. Observa-se que no ano de 2017 o Japão aparece com um percentual de 100%. **É crucial notar que este valor provavelmente reflete uma escassez de dados no dataset para 2017, com pouquíssimos jogos listados e/ou com dados de vendas apenas para o Japão, e não uma dominância real de 100% do mercado global.**
* **Crescimento das "Outras Regiões":** A fatia correspondente a "Outras Regiões" parece estável de 2009 à 2016, já em 2017 e 2020 sua participação foi muito baixa. **Similarmente, a baixa participação (ou a dominância de uma única região como NA em 2020) nesses anos mais recentes deve ser interpretada com cautela, pois o dataset `vgsales.csv` é conhecido por ser menos completo para lançamentos após aproximadamente 2016.** Uma investigação sobre o número de títulos disponíveis para esses anos no dataset é recomendada antes de tirar conclusões definitivas sobre o market share nesses períodos.

Essas tendências de market share são importantes para publicadoras e desenvolvedoras entenderem a distribuição geográfica de suas vendas e adaptarem suas estratégias de lançamento e marketing.

## Análise mais refinada do Market Share Regional ao longo dos anos
Como é possível ver nos dados da tabela **Contagem de jogos por ano (últimos anos)** é possível notar que a planilha carece de dados suficientes para os anos após 2016. Então, para uma análise mais refinada, é melhor excluirmos esses dados

```python
df_games_ate_2016 = df_games[df_games['year'] <= 2016].copy()

sales_by_year_region_filtrado = df_games_ate_2016.groupby('year')[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum()
sales_by_year_region_filtrado['total_annual_sales'] = sales_by_year_region_filtrado.sum(axis=1)
market_share_by_year_filtrado = sales_by_year_region_filtrado[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].apply(
    lambda x: (x / sales_by_year_region_filtrado['total_annual_sales']) * 100, axis=0
)

plt.figure(figsize=(17, 9))
market_share_by_year_filtrado.plot(kind='area', stacked=True, figsize=(17,9), colormap='viridis')
plt.title('Participação de Mercado (%) das Vendas Globais por Região (Até 2016)', fontsize=16)
plt.ylabel('Participação de Mercado (%)', fontsize=14)
plt.xlabel('Ano de Lançamento', fontsize=14)
plt.legend(title='Região', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print('\nParticipação de Mercado (%) nos últimos anos:')
print(market_share_by_year.tail(10).head(-2))
```
